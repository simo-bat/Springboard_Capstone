### Summary of trained models

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
path='/Crack_detection/Crack_detection/'
models_summary=pd.read_csv(os.path.join(path,'models_summary.csv'))
models_summary[['model','cnn_layers','dropout', 'count_params','train_recall', 'val_recall', 'test_recall', 'train_loss', 'val_loss', 'test_loss']].sort_values(by=['dropout','cnn_layers','count_params'])
```

#### Most imoprtant metric is recall = (true positive)/(true positive + false negative)

```python
for dp in np.sort(models_summary.dropout.unique()):
    ms_dp=models_summary.iloc[models_summary[models_summary.dropout==dp].index,:]
    ax1 = ms_dp.plot(kind='scatter', x='cnn_layers', y='train_recall', color='r', label='train', s=100)    
    ax2 = ms_dp.plot(kind='scatter', x='cnn_layers', y='val_recall', color='g', ax=ax1, label='validation', s=100)    
    ax3 = ms_dp.plot(kind='scatter', x='cnn_layers', y='test_recall', color='b', ax=ax1, label = 'test', s=100)
    plt.title('Recall for dropout = '+str(dp))
    plt.ylabel('Recall')
    plt.ylim(0.25,1.05)
    plt.legend(loc='best')
```
